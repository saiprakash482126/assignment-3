1.Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed
and average_of_vehicle.

sol:

class Vehicle:
    # Constructor (__init__) method
    def __init__(self, name_of_vehicle, max_speed, average_speed):

        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_speed = average_speed

 Creating an object (instance) of the 'Vehicle' class

my_vehicle = Vehicle(name_of_vehicle="Car", max_speed=200, average_speed=60)

3.What is multiple inheritance? Write a python code to demonstrate multiple inheritance.

sol:

Multiple inheritance is a feature in object-oriented programming where a class can inherit attributes and methods from more than one base class. In Python, a class can be derived from multiple classes by specifying them in the class definition.

example:

class Engine:
    def start(self):
        return "Engine started"

class ElectricMotor:
    def charge(self):
        return "Electric motor charging"

class HybridCar(Engine, ElectricMotor):
    def drive(self):
        return "Hybrid car is driving"

 Creating an object (instance) of the 'HybridCar' class

my_hybrid_car = HybridCar()

 Accessing methods from both base classes

print(my_hybrid_car.start())   # Output: Engine started
print(my_hybrid_car.charge())  # Output: Electric motor charging
print(my_hybrid_car.drive())   # Output: Hybrid car is driving

output:

Engine started

Electric motor charging

Hybrid car is driving


Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this
class.

sol:

class Student:
    # Constructor (__init__) method
    def __init__(self, name, age):
        # Instance variables
        self._name = name  # Note: Conventionally, a single leading underscore indicates a protected attribute
        self._age = age

    # Getter method for 'name'
    def get_name(self):
        return self._name

    # Setter method for 'name'
    def set_name(self, name):
        self._name = name

    # Getter method for 'age'
    def get_age(self):
        return self._age

    # Setter method for 'age'
    def set_age(self, age):
        if age > 0:
            self._age = age
        else:
            print("Invalid age value. Age must be greater than 0.")

 Creating an object (instance) of the 'Student' class

student1 = Student(name="Alice", age=20)

 Accessing attributes using getters

print("Name:", student1.get_name())  # Output: Name: Alice
print("Age:", student1.get_age())    # Output: Age: 20

 Modifying attributes using setters

student1.set_name("Bob")
student1.set_age(25)

 Accessing modified attributes using getters

print("Updated Name:", student1.get_name())  # Output: Updated Name: Bob
print("Updated Age:", student1.get_age())    # Output: Updated Age: 25

output:

Name: Alice

Age: 20

Updated Name: Bob

Updated Age: 25
